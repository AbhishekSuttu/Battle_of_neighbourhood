### function that extracts the category of the venue - borrow from the Foursquare lab.

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
IGnearby_venues = json_normalize(venues) # flatten JSON
# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
IGnearby_venues =IGnearby_venues.loc[:, filtered_columns]
# filter the category for each row
IGnearby_venues['venue.categories'] = IGnearby_venues.apply(get_category_type, axis=1)
# clean columns
IGnearby_venues.columns = [col.split(".")[-1] for col in IGnearby_venues.columns]

IGnearby_venues.shape
```
