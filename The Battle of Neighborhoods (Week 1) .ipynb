```python
# results display is hidden for report simplification 
results = requests.get(url).json()
results
```

```python
def get_category_type(row):
    try:
        categories_list = row['categories']
    except:
        categories_list = row['venue.categories']
        
    if len(categories_list) == 0:
        return None
    else:
        return categories_list[0]['name']
```

```python
venues = results['response']['groups'][0]['items']
IGnearby_venues = json_normalize(venues) # flatten JSON
# filter columns
filtered_columns = ['venue.name', 'venue.categories', 'venue.location.lat', 'venue.location.lng']
IGnearby_venues =IGnearby_venues.loc[:, filtered_columns]
# filter the category for each row
IGnearby_venues['venue.categories'] = IGnearby_venues.apply(get_category_type, axis=1)
# clean columns
IGnearby_venues.columns = [col.split(".")[-1] for col in IGnearby_venues.columns]

IGnearby_venues.shape
```

# Map of India with venues near residence place - for reference

```python
latitude=18.9387711
longitude=72.8353355

# create map of Singapore place  using latitude and longitude values
map_ig = folium.Map(location=[latitude, longitude], zoom_start=18)
# add markers to map
for lat, lng, label in zip(IGnearby_venues['lat'], IGnearby_venues['lng'], IGnearby_venues['name']):
    label = folium.Popup(label, parse_html=True)
    folium.RegularPolygonMarker(
        [lat, lng],
        number_of_sides=30,
        radius=7,
        popup=label,
        color='blue',
        fill_color='#0f0f0f',
        fill_opacity=0.6,
    ).add_to(map_ig)  
    
map_ig
```
